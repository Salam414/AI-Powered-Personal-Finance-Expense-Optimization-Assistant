```python
import pandas as pd
```

```python
df=pd.read_csv("Personal_Finance_Dataset_CLEANED.csv")
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df["Amount"] = df.apply(
    lambda row: -abs(row["Amount"]) if row["Type"] == "Expense" else abs(row["Amount"]),
    axis=1
)
```

```python
total_spending = abs(df[df["Type"] == "Expense"]["Amount"].sum())
total_spending
```

```python
category_totals = (
    df[df["Type"] == "Expense"]
    .groupby("Category")["Amount"]
    .sum()
    .abs()
)

category_totals
```

```python
needs_limit = 0.50 * total_spending
wants_limit = 0.30 * total_spending
savings_target = 0.20 * total_spending

needs_limit, wants_limit, savings_target
```

```python
needs_categories = ["Rent", "Food & Drink", "Utilities", "Transportation"]
wants_categories = ["Shopping", "Entertainment", "Travel"]
```

```python
needs_spending = category_totals[
    category_totals.index.isin(needs_categories)
].sum()

wants_spending = category_totals[
    category_totals.index.isin(wants_categories)
].sum()

needs_spending, wants_spending
```

```python
violations = {}

if needs_spending > needs_limit:
    violations["Needs"] = "Overspending"

if wants_spending > wants_limit:
    violations["Wants"] = "Overspending"

violations
```

```python
recommendations = []

if "Needs" in violations:
    recommendations.append(
        "Reduce essential expenses such as food, rent, or utilities."
    )

if "Wants" in violations:
    recommendations.append(
        "Limit non-essential spending like shopping or entertainment."
    )

if not recommendations:
    recommendations.append(
        "Your spending follows recommended budgeting rules. Keep it up."
    )

recommendations
```
