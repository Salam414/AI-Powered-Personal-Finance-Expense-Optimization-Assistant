```python
import pandas as pd
```

```python
df=pd.read_csv("personal_Finance_Dataset_processed.csv")
```

```python
df.head()
```

```python
df.isnull().sum() #if all 0s perfect no missing data
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
df["Amount"]=df.apply(lambda row:-abs(row["Amount"]) if row ["Type"]=="Expense" else abs(row["Amount"]),axis=1)
#this will tell if the amount is - so it is an expense if + it is an income
```

```python
df[["Type", "Amount"]].head()
```

```python
df.duplicated().sum() #ra7 yetla3 ljaweb 0 so no duplicate rows
```

```python
df["Amount"].describe()
```

```python
Q1 = df["Amount"].quantile(0.25)
Q3 = df["Amount"].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)
```

```python
outliers = ((df["Amount"] < lower_limit) | (df["Amount"] > upper_limit)).sum()
print("Number of outliers:", outliers)
```

```python
# Cap (limit) outliers within the lower and upper limits
df["Amount"] = df["Amount"].clip(lower=lower_limit, upper=upper_limit)

# Let's confirm the new limits
df["Amount"].describe()
```

```python
(df["Amount"]<lower_limit).sum(),
(df["Amount"]>upper_limit).sum()
```

```python
# Clean up text columns: strip spaces and standardize case
text_columns = ["Category", "Type", "Transaction Description"]

for col in text_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()

df.head()
```
