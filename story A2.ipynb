```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Personal_Finance_Dataset_CLEANED.csv")
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.describe()
```

```python
df["Type"].value_counts()
df["Category"].value_counts()
```

```python
summary = df.groupby("Type")["Amount"].sum()
summary
```

```python
summary.plot(kind="bar", title="Total Income vs Expenses")
plt.ylabel("Amount")
plt.show()
```

```python
df["Month"] = df["Date"].dt.to_period("M")

monthly_trend = df.groupby("Month")["Amount"].sum()
monthly_trend
```

```python
monthly_trend.plot(kind="line", title="Monthly Net Cash Flow")
plt.ylabel("Amount")
plt.xlabel("Month")
plt.show()
```

```python
expenses = df[df["Type"] == "expense"]

category_expense = expenses.groupby("Category")["Amount"].sum().abs()
category_expense
```

```python
category_expense.plot(kind="bar", title="Expenses by Category")
plt.ylabel("Total Expense")
plt.show()
```
